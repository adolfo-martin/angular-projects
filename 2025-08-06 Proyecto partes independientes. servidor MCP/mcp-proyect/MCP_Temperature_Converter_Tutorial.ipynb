```python
# Explore project structure
import os
from pathlib import Path

def show_tree(directory, prefix="", max_depth=3, current_depth=0):
    """Display directory tree structure"""
    if current_depth > max_depth:
        return
    
    directory = Path(directory)
    items = sorted([item for item in directory.iterdir() if not item.name.startswith('.')])
    
    for i, item in enumerate(items):
        is_last = i == len(items) - 1
        current_prefix = "└── " if is_last else "├── "
        print(f"{prefix}{current_prefix}{item.name}")
        
        if item.is_dir() and current_depth < max_depth:
            extension = "    " if is_last else "│   "
            show_tree(item, prefix + extension, max_depth, current_depth + 1)

print("📁 Project Structure:")
project_root = Path.cwd()
print(f"📂 {project_root.name}/")
show_tree(project_root)

print("\n📝 Key Files Explanation:")
print("├── pyproject.toml          - Project configuration and dependencies")
print("├── README.md               - Project documentation")  
print("├── src/                    - Source code directory")
print("│   └── temperature_converter_mcp/")
print("│       ├── __init__.py     - Main MCP server implementation")
print("│       └── __main__.py     - Entry point for module execution")
print("├── .vscode/")
print("│   └── mcp.json           - VS Code MCP configuration")
print("└── .github/")
print("    └── copilot-instructions.md - AI assistant guidelines")
```

# Section 3: Understanding MCP Concepts

## 3.1 What is a Model Context Protocol (MCP) Server?

An MCP server is a service that:
- **Exposes tools** that AI assistants can use
- **Follows a standard protocol** for communication
- **Runs as a separate process** connected via stdio or HTTP
- **Provides structured data** to AI models

## 3.2 Key MCP Components

### 🔧 Tools
Functions that the AI can call with specific parameters. In our case:
- `celsius_to_fahrenheit` - Convert °C to °F
- `fahrenheit_to_celsius` - Convert °F to °C

### 📡 Server
Handles incoming requests and manages tool execution

### 🔌 Transport
Communication method between AI and server (stdio, HTTP, etc.)

## 3.3 Temperature Conversion Formulas

Let's implement and test our conversion logic:

```python
# Temperature Conversion Implementation
class TemperatureConverter:
    """Educational temperature converter with detailed explanations"""
    
    @staticmethod
    def celsius_to_fahrenheit(celsius: float) -> float:
        """
        Convert Celsius to Fahrenheit
        Formula: °F = (°C × 9/5) + 32
        
        Why this formula works:
        - Celsius scale: 0°C = freezing, 100°C = boiling (100 degree range)
        - Fahrenheit scale: 32°F = freezing, 212°F = boiling (180 degree range)
        - Ratio: 180/100 = 9/5
        - Offset: Fahrenheit freezing point is 32°F
        """
        fahrenheit = (celsius * 9/5) + 32
        return fahrenheit
    
    @staticmethod 
    def fahrenheit_to_celsius(fahrenheit: float) -> float:
        """
        Convert Fahrenheit to Celsius
        Formula: °C = (°F - 32) × 5/9
        
        This is the inverse of the Celsius to Fahrenheit formula:
        - First subtract the offset (32)
        - Then scale by the inverse ratio (5/9)
        """
        celsius = (fahrenheit - 32) * 5/9
        return celsius

# Test our conversion functions
converter = TemperatureConverter()

print("🧪 Testing Temperature Conversions:")
print("\n--- Celsius to Fahrenheit ---")
test_celsius = [0, 25, 37, 100, -40]
for c in test_celsius:
    f = converter.celsius_to_fahrenheit(c)
    print(f"{c}°C = {f}°F")

print("\n--- Fahrenheit to Celsius ---")  
test_fahrenheit = [32, 77, 98.6, 212, -40]
for f in test_fahrenheit:
    c = converter.fahrenheit_to_celsius(f)
    print(f"{f}°F = {c:.1f}°C")

print("\n🔄 Round-trip Test (should return original values):")
original_temp = 25
converted = converter.celsius_to_fahrenheit(original_temp)
back_converted = converter.fahrenheit_to_celsius(converted)
print(f"Original: {original_temp}°C → {converted}°F → {back_converted:.1f}°C")
```
