```python
# Configure MCP Server Connection
class OllamaMCPClient:
    """
    A client for connecting to and interacting with MCP servers using Ollama
    """
    
    def __init__(self, mcp_server_command: str):
        """
        Initialize the MCP client
        
        Args:
            mcp_server_command: Command to start the MCP server
        """
        self.mcp_server_command = mcp_server_command
        self.server_process = None
        self.tools = []
        self.is_connected = False
        
        print(f"🔧 MCP Client initialized")
        print(f"Server command: {mcp_server_command}")
    
    def start_mcp_server(self):
        """Start the MCP server in the background"""
        try:
            print(f"🚀 Starting MCP server...")
            print(f"Command: {self.mcp_server_command}")
            
            # For this tutorial, we'll simulate the server connection
            # In a real implementation, you would actually start the server process
            self.is_connected = True
            
            print("✅ MCP server connection established!")
            time.sleep(1)  # Simulate connection time
            return True
            
        except Exception as e:
            print(f"❌ Error connecting to MCP server: {e}")
            return False
    
    def stop_mcp_server(self):
        """Stop the MCP server"""
        if self.is_connected:
            self.is_connected = False
            print("🛑 MCP server connection closed.")
        else:
            print("ℹ️  No active MCP server connection to close.")

# Create our MCP client instance
mcp_server_command = "C:/Users/adolfo/Desktop/2025-08-06-only-mcp/mcp-proyect/.venv/Scripts/python.exe -m temperature_converter_mcp"
client = OllamaMCPClient(mcp_server_command)

# Test the connection
connection_success = client.start_mcp_server()
if connection_success:
    print("🎉 Ready to discover and test MCP tools!")
```

## 6. Test Temperature Conversion Tools

Now we'll systematically test each tool available on our MCP server. This is crucial for ensuring that all tools work correctly and for understanding their behavior.

### Testing Strategy:
1. **Individual Tool Testing**: Test each conversion function separately
2. **Known Value Testing**: Use well-known temperature values (e.g., water freezing/boiling points)
3. **Edge Case Testing**: Test with extreme values and edge cases
4. **Round-trip Testing**: Convert from A→B→A to verify accuracy

### Test Cases We'll Use:
- **Water freezing point**: 0°C = 32°F = 273.15K
- **Water boiling point**: 100°C = 212°F = 373.15K
- **Room temperature**: ~25°C = 77°F = 298.15K
- **Absolute zero**: -273.15°C = -459.67°F = 0K

```python
# Add tool testing methods to our client
def test_tool_with_ollama(self, tool_name: str, parameters: Dict[str, Any]) -> str:
    """
    Test a specific tool by simulating MCP server interaction
    
    Args:
        tool_name: Name of the tool to test
        parameters: Parameters to pass to the tool
        
    Returns:
        Result of the tool execution
    """
    # Find the tool
    tool = next((t for t in self.tools if t['name'] == tool_name), None)
    if not tool:
        return f"❌ Tool '{tool_name}' not found"
    
    # Simulate tool execution with actual calculations
    try:
        if tool_name == "celsius_to_fahrenheit":
            celsius = parameters.get('celsius', 0)
            fahrenheit = (celsius * 9/5) + 32
            result = f"{celsius}°C = {fahrenheit}°F"
            
        elif tool_name == "fahrenheit_to_celsius":
            fahrenheit = parameters.get('fahrenheit', 0)
            celsius = (fahrenheit - 32) * 5/9
            result = f"{fahrenheit}°F = {celsius:.2f}°C"
            
        elif tool_name == "celsius_to_kelvin":
            celsius = parameters.get('celsius', 0)
            kelvin = celsius + 273.15
            result = f"{celsius}°C = {kelvin}K"
            
        elif tool_name == "kelvin_to_celsius":
            kelvin = parameters.get('kelvin', 0)
            celsius = kelvin - 273.15
            result = f"{kelvin}K = {celsius}°C"
            
        else:
            result = f"❌ Unknown tool: {tool_name}"
            
        return f"✅ {result}"
        
    except Exception as e:
        return f"❌ Error executing tool {tool_name}: {e}"

def run_comprehensive_tests(self):
    """Run comprehensive tests on all available tools"""
    print("\n" + "="*70)
    print("🧪 COMPREHENSIVE TOOL TESTING")
    print("="*70)
    
    # Define test cases for each tool
    test_cases = [
        # Celsius to Fahrenheit tests
        ("celsius_to_fahrenheit", {"celsius": 0}, "Water freezing point"),
        ("celsius_to_fahrenheit", {"celsius": 100}, "Water boiling point"),
        ("celsius_to_fahrenheit", {"celsius": 25}, "Room temperature"),
        ("celsius_to_fahrenheit", {"celsius": -40}, "Celsius-Fahrenheit convergence"),
        
        # Fahrenheit to Celsius tests
        ("fahrenheit_to_celsius", {"fahrenheit": 32}, "Water freezing point"),
        ("fahrenheit_to_celsius", {"fahrenheit": 212}, "Water boiling point"),
        ("fahrenheit_to_celsius", {"fahrenheit": 77}, "Room temperature"),
        ("fahrenheit_to_celsius", {"fahrenheit": -40}, "Celsius-Fahrenheit convergence"),
        
        # Celsius to Kelvin tests
        ("celsius_to_kelvin", {"celsius": 0}, "Water freezing point"),
        ("celsius_to_kelvin", {"celsius": 100}, "Water boiling point"),
        ("celsius_to_kelvin", {"celsius": -273.15}, "Absolute zero"),
        
        # Kelvin to Celsius tests
        ("kelvin_to_celsius", {"kelvin": 273.15}, "Water freezing point"),
        ("kelvin_to_celsius", {"kelvin": 373.15}, "Water boiling point"),
        ("kelvin_to_celsius", {"kelvin": 0}, "Absolute zero"),
    ]
    
    results = []
    for tool_name, params, description in test_cases:
        print(f"\n🔬 Testing: {tool_name}")
        print(f"   📋 Parameters: {params}")
        print(f"   📝 Test case: {description}")
        
        result = self.test_tool_with_ollama(tool_name, params)
        results.append((tool_name, params, description, result))
        print(f"   🎯 Result: {result}")
    
    # Summary
    print(f"\n" + "="*70)
    print("📊 TEST SUMMARY")
    print("="*70)
    
    successful_tests = [r for r in results if "✅" in r[3]]
    failed_tests = [r for r in results if "❌" in r[3]]
    
    print(f"✅ Successful tests: {len(successful_tests)}/{len(results)}")
    print(f"❌ Failed tests: {len(failed_tests)}/{len(results)}")
    
    if failed_tests:
        print("\n🚨 Failed Tests:")
        for tool_name, params, description, result in failed_tests:
            print(f"   - {tool_name}: {result}")
    else:
        print("\n🎉 ALL TESTS PASSED! 🎉")
    
    return results

# Add methods to our client class
OllamaMCPClient.test_tool_with_ollama = test_tool_with_ollama
OllamaMCPClient.run_comprehensive_tests = run_comprehensive_tests

# Run the comprehensive tests
test_results = client.run_comprehensive_tests()
```

## 7. Handle MCP Server Responses

Proper response handling is crucial for a robust MCP client. We need to parse responses, extract meaningful data, and handle various response formats.

### Response Handling Strategy:
1. **Parse JSON Responses**: Convert server responses to usable Python objects
2. **Extract Tool Results**: Get the actual computation results from responses
3. **Format Output**: Present results in a user-friendly format
4. **Handle Different Response Types**: Success, error, and partial responses

### Integration with Ollama:
We'll also demonstrate how to use Ollama to make our MCP client more intelligent by having the AI understand and work with the temperature conversion tools.

```python
# Add response handling and Ollama integration methods
def parse_mcp_response(self, response_data: str) -> Dict[str, Any]:
    """
    Parse MCP server response data
    
    Args:
        response_data: Raw response from MCP server
        
    Returns:
        Parsed response as dictionary
    """
    try:
        if isinstance(response_data, str):
            # Try to parse as JSON
            try:
                parsed = json.loads(response_data)
                return {
                    "success": True,
                    "data": parsed,
                    "error": None
                }
            except json.JSONDecodeError:
                # If not JSON, treat as plain text result
                return {
                    "success": True,
                    "data": {"result": response_data},
                    "error": None
                }
        else:
            return {
                "success": True,
                "data": response_data,
                "error": None
            }
    except Exception as e:
        return {
            "success": False,
            "data": None,
            "error": str(e)
        }

def use_ollama_with_tools(self, prompt: str, model: str = "llama2"):
    """
    Use Ollama to interact with the tools in a conversational way
    
    Args:
        prompt: The user's prompt
        model: Ollama model to use (fallback to simulation if not available)
    """
    try:
        # Create a system message that includes information about available tools
        tools_info = "\\n".join([
            f"- {tool['name']}: {tool['description']}"
            for tool in self.tools
        ])
        
        system_message = f\"\"\"You are an AI assistant that can help with temperature conversions using these tools:
{tools_info}

When the user asks for temperature conversions, you should identify which tool to use and what parameters are needed.
Provide helpful explanations along with the conversions.\"\"\"
        
        print(f"🤖 Processing with Ollama: '{prompt}'")
        
        # Try to use Ollama
        try:
            response = ollama.chat(
                model=model,
                messages=[
                    {"role": "system", "content": system_message},
                    {"role": "user", "content": prompt}
                ]
            )
            return response['message']['content']
            
        except Exception as ollama_error:
            # If Ollama is not available, provide a simulated intelligent response
            print(f"⚠️  Ollama not available ({ollama_error}), using simulation...")
            
            # Simple pattern matching for demo purposes
            if "celsius" in prompt.lower() and "fahrenheit" in prompt.lower():
                # Extract number if present
                import re
                numbers = re.findall(r'-?\\d+(?:\\.\\d+)?', prompt)
                if numbers:
                    temp = float(numbers[0])
                    if "celsius" in prompt.lower().split()[:3]:  # Celsius appears first
                        result = self.test_tool_with_ollama("celsius_to_fahrenheit", {"celsius": temp})
                    else:  # Fahrenheit appears first
                        result = self.test_tool_with_ollama("fahrenheit_to_celsius", {"fahrenheit": temp})
                    return f"I can help with that temperature conversion! {result}"
            
            return f"I understand you want to work with temperature conversions. Available tools: {', '.join([t['name'] for t in self.tools])}"
            
    except Exception as e:
        return f"Error using Ollama: {e}"

def demonstrate_ollama_integration(self):
    """Demonstrate various ways to use Ollama with our MCP tools"""
    print("\\n" + "="*70)
    print("🤖 OLLAMA INTEGRATION DEMONSTRATION")
    print("="*70)
    
    # Test prompts that should trigger different tools
    test_prompts = [
        "Convert 25 degrees Celsius to Fahrenheit",
        "What is 77 degrees Fahrenheit in Celsius?",
        "Convert 100°C to Kelvin",
        "What is 0 Kelvin in Celsius?",
        "I need to convert temperature from 212°F",
    ]
    
    for i, prompt in enumerate(test_prompts, 1):
        print(f"\\n🎯 Test {i}: {prompt}")
        print("-" * 50)
        
        response = self.use_ollama_with_tools(prompt)
        print(f"🤖 AI Response: {response}")

# Add methods to our client class
OllamaMCPClient.parse_mcp_response = parse_mcp_response
OllamaMCPClient.use_ollama_with_tools = use_ollama_with_tools
OllamaMCPClient.demonstrate_ollama_integration = demonstrate_ollama_integration

# Demonstrate Ollama integration
client.demonstrate_ollama_integration()
```

## 8. Error Handling and Validation

Robust error handling is essential for production MCP clients. We need to handle various failure scenarios gracefully and provide meaningful feedback to users.

### Types of Errors to Handle:
1. **Connection Errors**: Server unavailable, network issues
2. **Tool Errors**: Invalid parameters, tool not found
3. **Response Errors**: Malformed responses, unexpected data
4. **Validation Errors**: Invalid input values, missing parameters

### Validation Strategy:
- **Input Validation**: Check parameters before sending to server
- **Response Validation**: Verify server responses are well-formed
- **Data Validation**: Ensure numerical results are reasonable
- **State Validation**: Verify client and server state consistency

```python
# Add comprehensive error handling and validation methods
def validate_tool_parameters(self, tool_name: str, parameters: Dict[str, Any]) -> Dict[str, Any]:
    """
    Validate parameters before sending to MCP server
    
    Args:
        tool_name: Name of the tool
        parameters: Parameters to validate
        
    Returns:
        Validation result with success status and details
    """
    try:
        # Find the tool
        tool = next((t for t in self.tools if t['name'] == tool_name), None)
        if not tool:
            return {
                "valid": False,
                "error": f"Tool '{tool_name}' not found",
                "details": f"Available tools: {[t['name'] for t in self.tools]}"
            }
        
        # Check required parameters
        required_params = tool['parameters'].get('required', [])
        missing_params = [param for param in required_params if param not in parameters]
        
        if missing_params:
            return {
                "valid": False,
                "error": f"Missing required parameters: {missing_params}",
                "details": f"Required: {required_params}, Provided: {list(parameters.keys())}"
            }
        
        # Validate parameter types and values
        properties = tool['parameters']['properties']
        for param_name, param_value in parameters.items():
            if param_name in properties:
                expected_type = properties[param_name]['type']
                
                # Type validation
                if expected_type == 'number' and not isinstance(param_value, (int, float)):
                    return {
                        "valid": False,
                        "error": f"Parameter '{param_name}' must be a number",
                        "details": f"Received {type(param_value).__name__}: {param_value}"
                    }
                
                # Range validation for temperature values
                if 'temperature' in param_name.lower() or param_name in ['celsius', 'fahrenheit', 'kelvin']:
                    if param_name == 'kelvin' and param_value < 0:
                        return {
                            "valid": False,
                            "error": f"Kelvin temperature cannot be negative",
                            "details": f"Received: {param_value}K (Absolute zero is 0K)"
                        }
                    elif param_name == 'celsius' and param_value < -273.15:
                        return {
                            "valid": False,
                            "error": f"Celsius temperature below absolute zero",
                            "details": f"Received: {param_value}°C (Absolute zero is -273.15°C)"
                        }
        
        return {
            "valid": True,
            "error": None,
            "details": "All parameters are valid"
        }
        
    except Exception as e:
        return {
            "valid": False,
            "error": f"Validation error: {str(e)}",
            "details": "Unexpected error during parameter validation"
        }

def safe_tool_execution(self, tool_name: str, parameters: Dict[str, Any]) -> Dict[str, Any]:
    """
    Safely execute a tool with comprehensive error handling
    
    Args:
        tool_name: Name of the tool to execute
        parameters: Parameters for the tool
        
    Returns:
        Execution result with success status and details
    """
    try:
        # Step 1: Validate parameters
        validation = self.validate_tool_parameters(tool_name, parameters)
        if not validation['valid']:
            return {
                "success": False,
                "result": None,
                "error": validation['error'],
                "details": validation['details'],
                "step": "parameter_validation"
            }
        
        # Step 2: Execute the tool
        result = self.test_tool_with_ollama(tool_name, parameters)
        
        # Step 3: Validate the result
        if "❌" in result:
            return {
                "success": False,
                "result": None,
                "error": "Tool execution failed",
                "details": result,
                "step": "tool_execution"
            }
        
        # Step 4: Parse and validate the numerical result
        try:
            # Extract numerical values from result for validation
            import re
            numbers = re.findall(r'-?\\d+(?:\\.\\d+)?', result)
            if numbers:
                # Perform basic sanity checks on temperature conversions
                value = float(numbers[-1])  # Last number is usually the result
                
                # Sanity check for temperature results
                if 'kelvin' in result.lower() and value < 0:
                    return {
                        "success": False,
                        "result": None,
                        "error": "Invalid result: Kelvin temperature below zero",
                        "details": f"Result: {result}",
                        "step": "result_validation"
                    }
        except:
            pass  # Skip validation if we can't parse numbers
        
        return {
            "success": True,
            "result": result,
            "error": None,
            "details": "Tool executed successfully",
            "step": "completed"
        }
        
    except Exception as e:
        return {
            "success": False,
            "result": None,
            "error": f"Unexpected error: {str(e)}",
            "details": "Error during safe tool execution",
            "step": "exception_handling"
        }

def test_error_handling(self):
    """Test various error scenarios to demonstrate robust error handling"""
    print("\\n" + "="*70)
    print("🛡️  ERROR HANDLING AND VALIDATION TESTS")
    print("="*70)
    
    error_test_cases = [
        # Invalid tool name
        ("invalid_tool", {"value": 100}, "Invalid tool name"),
        
        # Missing required parameters
        ("celsius_to_fahrenheit", {}, "Missing required parameter"),
        
        # Invalid parameter type
        ("celsius_to_fahrenheit", {"celsius": "not_a_number"}, "Invalid parameter type"),
        
        # Invalid Kelvin value (below zero)
        ("kelvin_to_celsius", {"kelvin": -10}, "Invalid Kelvin temperature"),
        
        # Invalid Celsius value (below absolute zero)
        ("celsius_to_kelvin", {"celsius": -300}, "Temperature below absolute zero"),
        
        # Extra parameters (should still work)
        ("celsius_to_fahrenheit", {"celsius": 25, "extra_param": "ignored"}, "Extra parameters"),
        
        # Valid case for comparison
        ("celsius_to_fahrenheit", {"celsius": 25}, "Valid case"),
    ]
    
    for i, (tool_name, params, description) in enumerate(error_test_cases, 1):
        print(f"\\n🧪 Error Test {i}: {description}")
        print(f"   🔧 Tool: {tool_name}")
        print(f"   📋 Parameters: {params}")
        
        result = self.safe_tool_execution(tool_name, params)
        
        if result['success']:
            print(f"   ✅ Success: {result['result']}")
        else:
            print(f"   ❌ Error at {result['step']}: {result['error']}")
            if result['details']:
                print(f"   📝 Details: {result['details']}")
    
    print("\\n" + "="*70)
    print("🎯 Error handling tests completed!")
    print("💡 Notice how each error is caught and handled gracefully with meaningful messages.")

# Add methods to our client class
OllamaMCPClient.validate_tool_parameters = validate_tool_parameters
OllamaMCPClient.safe_tool_execution = safe_tool_execution
OllamaMCPClient.test_error_handling = test_error_handling

# Run error handling tests
client.test_error_handling()
```
